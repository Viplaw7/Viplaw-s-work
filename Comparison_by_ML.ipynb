Comparison of power with ML By linear Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Define temperature values
temperature_values = np.array([25, 30, 35, 40, 45, 50]).reshape(-1, 1)

# Define corresponding power and efficiency values (example dataset)
power_values = np.array([1.83, 1.825, 1.8125, 1.78, 1.775, 1.77])
efficiency_values = 10 * power_values  # Simulated efficiency

# Create a DataFrame
data = pd.DataFrame({'Temperature': temperature_values.flatten(),
                     'Power': power_values,
                     'Efficiency': efficiency_values})

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    data[['Temperature']], data['Power'], test_size=0.2, random_state=42)

# Train Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test)

# Model Evaluation
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse:.5f}')
print(f'R Score: {r2:.5f}')

# Visualization
plt.scatter(X_test, y_test, color='blue', label='Actual Power')
plt.plot(X_test, y_pred, color='red', linewidth=2, label='Predicted Power')
plt.xlabel('Temperature (C)')
plt.ylabel('Power')
plt.title('Linear Regression: Power vs Temperature')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Define temperature values
temperature_values = np.array([25, 30, 35, 40, 45, 50]).reshape(-1, 1)

# Define corresponding power, efficiency, Isc, Voc, Imax, and Vmax values (example dataset)
power_values = np.array([1.83, 1.825, 1.8125, 1.78, 1.775, 1.77])
efficiency_values = 10 * power_values  # Simulated efficiency
isc_values = np.array([2.54, 2.55, 2.565, 2.58, 2.595, 2.61])
voc_values = np.array([0.77, 0.78, 0.79, 0.80, 0.809, 0.818])
imax_values = np.array([2.74, 2.78, 2.775, 2.67, 2.675, 2.73])
vmax_values = np.array([0.64, 0.63, 0.625, 0.66, 0.655, 0.65])

# Create a DataFrame
data = pd.DataFrame({'Temperature': temperature_values.flatten(),
                     'Power': power_values,
                     'Efficiency': efficiency_values,
                     'Isc': isc_values,
                     'Voc': voc_values,
                     'Imax': imax_values,
                     'Vmax': vmax_values})

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    data[['Temperature', 'Isc', 'Voc', 'Imax', 'Vmax']], data['Power'], test_size=0.2, random_state=42)

# Train Random Forest Regressor model
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test)

# Model Evaluation
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse:.5f}')
print(f'R Score: {r2:.5f}')

# Visualization
plt.scatter(y_test, y_pred, color='blue', label='Actual vs Predicted Power')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], linestyle='--', color='red', label='Perfect Fit')
plt.xlabel('Actual Power')
plt.ylabel('Predicted Power')
plt.title('Random Forest Regression: Predicted vs Actual Power')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Define temperature values
temperature_values = np.array([25, 30, 35, 40, 45, 50]).reshape(-1, 1)

# Define corresponding efficiency, Isc, and Voc values (example dataset)
efficiency_values = np.array([18.3, 18.25, 18.125, 17.8, 17.75, 17.7])
isc_values = np.array([2.54, 2.55, 2.565, 2.58, 2.595, 2.61])
voc_values = np.array([0.77, 0.78, 0.79, 0.80, 0.809, 0.818])

# Create a DataFrame
data = pd.DataFrame({'Temperature': temperature_values.flatten(),
                     'Efficiency': efficiency_values,
                     'Isc': isc_values,
                     'Voc': voc_values})

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    data[['Temperature']], data[['Efficiency', 'Isc', 'Voc']], test_size=0.2, random_state=42)

# Train Random Forest Regressor models
models = {}
predictions = {}
metrics = {}
for target in ['Efficiency', 'Isc', 'Voc']:
    model = RandomForestRegressor(n_estimators=100, random_state=42)
    model.fit(X_train, y_train[target])
    models[target] = model
    predictions[target] = model.predict(X_test)
    mse = mean_squared_error(y_test[target], predictions[target])
    r2 = r2_score(y_test[target], predictions[target])
    metrics[target] = (mse, r2)
    print(f'{target} - Mean Squared Error: {mse:.5f}, R Score: {r2:.5f}')

# Visualization
fig, axes = plt.subplots(3, 1, figsize=(8, 12))
for i, target in enumerate(['Efficiency', 'Isc', 'Voc']):
    axes[i].scatter(y_test[target], predictions[target], color='blue', label=f'Actual vs Predicted {target}')
    axes[i].plot([min(y_test[target]), max(y_test[target])], [min(y_test[target]), max(y_test[target])], linestyle='--', color='red', label='Perfect Fit')
    axes[i].set_xlabel(f'Actual {target}')
    axes[i].set_ylabel(f'Predicted {target}')
    axes[i].set_title(f'Random Forest Regression: Predicted vs Actual {target}')
    axes[i].legend()

plt.tight_layout()
plt.show()
```
